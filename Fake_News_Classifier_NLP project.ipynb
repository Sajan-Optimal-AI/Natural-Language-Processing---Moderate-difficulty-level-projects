```python
# Import the data
import pandas as pd
Data = pd.read_csv('/content/drive/MyDrive/fake-news/train.csv')
Data.head()
```

```python
# check if any null value:
Data[Data.isnull().any(axis=1)]
```

```python
# Remove the records where NaN presents
Data = Data.dropna()
Data.shape
```

```python
# Reset the index's which were misaligned due to the removal of null records
Data.reset_index(inplace=True)
Data.head()
```

```python
# Get the dependent and independent features
Independent = Data.drop('label', axis=1)
Dependent = Data['label']
Dependent.head()
```

```python
# Summarize all the text in 'text' features  in Independent:
messages = Independent.copy()
messages['text'][6]
```

```python
import nltk
nltk.download('all')
```

```python
# Text screening and preprocessing using lemmatize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
Lemmatizer = WordNetLemmatizer()
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['text'][i])
    review = review.lower()
    review = review.split()
    
    review = [Lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus[3]
```

```python
# Convert corpus into Tf-Idf vectors
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_v=TfidfVectorizer(max_features=5000,ngram_range=(1,3))
X=tfidf_v.fit_transform(corpus).toarray()
X.shape
```

```python
# Extract 10 feature names
tfidf_v.get_feature_names()[:10]
```

```python
# Divide the dataset into Train and Test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Dependent, test_size=0.33, random_state=0)
```

```python
# Details about X_train:
X_train_df = pd.DataFrame(X_train, columns=tfidf_v.get_feature_names())
X_train_df.head()
```

```python
#Define Naive bayes model:
from sklearn.naive_bayes import MultinomialNB
classifier=MultinomialNB()
from sklearn import metrics
import numpy as np
```

```python
#Evaluate the model
classifier.fit(X_train, y_train)
pred = classifier.predict(X_test)
score = metrics.accuracy_score(y_test, pred)
print("accuracy:   %0.3f" % score)
print(metrics.confusion_matrix(y_test, pred))
print(metrics.classification_report(y_test,pred))
```

```python
# Hyperparameter tuning : To check for better performance but unfortunately gives same performance.
classifier=MultinomialNB(alpha=0.1)
```

```python
previous_score=0
for alpha in np.arange(0,1,0.1):
    sub_classifier=MultinomialNB(alpha=alpha)
    sub_classifier.fit(X_train,y_train)
    y_pred=sub_classifier.predict(X_test)
    score = metrics.accuracy_score(y_test, y_pred)
    if score>previous_score:
        classifier=sub_classifier
    print("Alpha: {}, Score : {}".format(alpha,score))
```
